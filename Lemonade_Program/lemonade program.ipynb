```python
import pandas as pd
```

```python
import tensorflow as tf
```

```python
# 파일로부터 데이터 읽어오기
파일경로 = 'https://raw.githubusercontent.com/sxxngii/AI2024/main/lemonade.csv'
레모네이드 = pd.read_csv(파일경로)
```

```python
# 변수 안에 어떤 데이터가 있는지 확인 (데이터 모양으로 확인)
print(레모네이드.shape)
```

```python
# 칼럼 이름 출력
print(레모네이드.columns)
```

index(['온도', '판매량'], dtype='object')

```python
독립 = 레모네이드[['온도']]
종속 = 레모네이드[['판매량']]
print(독립.shape, 종속.shape)
```

```python
# 실제데이터 출력

레모네이드.head()
```

```python
# 모델 구성
X = tf.keras.layers.Input(shape=[1])  # tf.keras.layers.input -> tf.keras.layers.Input
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')  # model.compil -> model.compile
```

```python
# 모델 학습
model.fit(독립, 종속, epochs=1000)
```

```python
# 예측
print("predictions :", model.predict([[15]]))
```
