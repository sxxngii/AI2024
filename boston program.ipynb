```python
#라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 1. 과거의 데이터를 준비합니다.
파일경로 = 'https://raw.githubusercontent.com/sxxngii/AI2024/main/boston.csv'
보스턴 = pd.read_csv(파일경로)
print(보스턴.columns)
보스턴.head()
```

```python
# 독립변수, 종속변수 분리
독립 = 보스턴[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]  # 쉼표 추가 및 'istat'을 'lstat'으로 수정
종속 = 보스턴[['medv']]
print(독립.shape, 종속.shape)  # shapem -> shape로 수정
```

```python
# 2. 모델의 구조를 만듭니다.
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 3. 데이터로 모델을 학습(FIT) 합니다.
model.fit(독립, 종속, epochs=1000, verbose=1)  # epochs 값 증가 및 verbose 추가
# model.fit(독립, 종속, epochs=10)
```

```python
# 4. 모델을 이용합니다.
print(model.predict(독립[5:10]))
# 종속변수 확인
print(종속[5:10])
```

```python
# 모델의 수식 확인
print(model.get_weights())
```
